### StatsModels 모듈을 사용한 선형회귀분석

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.stats.api as sms
```

```python
import pandas as pd
# 독립변수
X = pd.DataFrame(airbnb, columns = ["체크인시간", "의사소통", "청결도", "위치", "가격", "호스트별점(슈퍼호스트)", "후기", "호스트응답률"])
# 종속변수
Y = pd.DataFrame(airbnb, columns = ["만족도"]) 
```

```python
# 데이터 준비
dfX0 = pd.DataFrame(airbnb, columns = ["체크인시간", "의사소통", "청결도", "위치", "가격", "호스트별점(슈퍼호스트)", "후기", "호스트응답률"]) # 독립변수
dfX = sm.add_constant(dfX0) # 자동으로 상수항을 만들어주지 않기 때문에 사용자가 add_constant 명령으로 상수항을 추가해야 한다.
dfy = pd.DataFrame(airbnb, columns = ["만족도"]) # 종속변수
```

```python
# model_airbnb라는 이름으로 OLS 모델 구조화
model_airbnb = sm.OLS(dfy, dfX) # (종속변수, 독립변수)

# OLS 회귀모형 피팅
result_airbnb = model_airbnb.fit()

# 회귀분석 결과 요약
print(result_airbnb.summary())
```

```python
dfy_new = result_airbnb.predict(dfX)

plt.scatter(dfy, dfy_new)
plt.xlabel(u"실제 만족도")
plt.ylabel(u"만족도 예측치")
plt.show()
```

### 데이터 입력

```python
# 후기 안좋은 숙소
f = open("badreview.txt", 'r', encoding='euc-kr')
message_of_guests = f.read()
print(message_of_guests)
f.close()
```

```python
# 후기 좋은 숙소
f = open( "AirBnB_reviews.txt", "r")
message_of_ab = f.read()
print(message_of_ab)
f.close()
```

```python
# 후기 안좋은 숙소 전처리
message_of_guests = message_of_guests.split('\n\n')
message_of_guests
```

```python
# 후기 좋은 숙소 전처리
message_of_ab = message_of_ab.split('\n\n')
message_of_ab
```

```python
# 후기 안좋은 숙소 빈도 계산
tf_vectorizer = CountVectorizer(min_df = 0.001, token_pattern=r'\w{2,}')

tf_corpus1 = tf_vectorizer.fit(message_of_guests)
tf_bow1 = tf_vectorizer.fit_transform(message_of_guests)
tf_bow1
```

```python
# 좋은 후기 빈도 계산
tf_vectorizer = CountVectorizer(min_df = 0.001, token_pattern=r'\w{2,}')

tf_corpus2 = tf_vectorizer.fit(message_of_ab)
tf_bow2 = tf_vectorizer.fit_transform(message_of_ab)
tf_bow2
```

```python
# 후기 안좋은 숙소 
print(len(tf_corpus1.vocabulary_))
print(len(tf_corpus1.get_feature_names()))
tf_corpus1.get_feature_names()
```

```python
# 후기 좋은 숙소
print(len(tf_corpus2.vocabulary_))
print(len(tf_corpus2.get_feature_names()))
tf_corpus2.get_feature_names()
```

```python
# 후기 안좋은 숙소
# frequency count
tf_word_sum1 = tf_bow1.toarray().sum(axis=0)
tf_word_name1 = tf_corpus1.get_feature_names()
tf_word_dict1 = {}

for i in range(len(tf_word_sum1)):
    tf_word_dict1[tf_word_name1[i]] = tf_word_sum1[i]
    
print(tf_corpus1.get_feature_names())
tf_word_sum1
```

```python
# 후기 좋은 숙소
# frequency count
tf_word_sum2 = tf_bow2.toarray().sum(axis=0)
tf_word_name2 = tf_corpus2.get_feature_names()
tf_word_dict2 = {}

for i in range(len(tf_word_sum2)):
    tf_word_dict2[tf_word_name2[i]] = tf_word_sum2[i]
    
print(tf_corpus2.get_feature_names())
tf_word_sum2
```

## 5. 분석 결과

### 회귀분석 분석결과
- Prob(F-statistic)값을 통해 회귀분석 모형이 유의미한지를 알 수 있다. 위 회귀 모형의 F-statistic값은 1.16e-12로, 0.05이하이기 때문에 통계적으로 유의하다.
- R-square은 결정계수로, 위 회귀분석의 정확도를 파악할 수 있다. 이 분석의 R-square 값은 0.764로 이 분석은 76%의 설명력을 가진다.
- P-value가 0.05미만인 변수들을 확인하였을 때 ‘청결도’, ‘후기’, ‘호스트 응답률’이 0.05 미만인 변수들로, 이 세 가지의 변수들만 유효하며, 만족도에 영향을 준다.
- 유효한 세 가지의 변수들 중 Coefficient 계수를 통해 영향력을 파악한다. ‘호스트 응답률’의 Coef 계수는 0.386, ‘청결도’의 Coef 계수는 0.2752, 그리고 ‘후기’의 Coef계수는 0.1962로, ‘호스트 응답률’, ‘청결도’, ‘후기’ 순서로 만족도에 영향을 미치는 정도가 크다.

### 그래프 분석 결과
- 회귀 분석에 대한 그래프 분석 결과, 독립 변수와 종속 변수는 선형관계가 있으며, 서로 양의 상관관계를 가진다.

### 좋은 후기 분석 결과
- 좋은 후기들의 워드 클라우드 분석 결과, 좋은 후기들 중에서는 ‘정말’, ‘너무’,’아주’ 같은 감탄사가 가장 빈번하게 쓰였다.
- 그 뒤를 이어 ‘좋은’,’좋았습니다’,’좋았어요’라는 긍정적인 답변이 많이 쓰였음을 확인할 수 있다.
- 이유에 해당하는 단어들 중에서는 ‘깨끗하고’,’깔끔하고’ 라는 단어가 많이 쓰였는데, 이는 청결도를 상징하는 단어가 빈번히 사용되었다고 할 수 있다.

### 좋은 후기 워드클라우드 시사점
- 감탄사와 ‘좋았습니다’ 의미의 단어를 제외하면 좋은 후기를 작성하게 되는 중요 요인은 청결도가 차지한다고 볼 수 있다. 이는 위 회귀 분석의 결과와도 어느 정도 일치한다고 볼 수 있다.
- 후기 또한 회귀 분석 결과 만족도에 중요한 변수이기 때문에, 좋은 후기를 얻는 것이 만족도를 높이는 방법이 될 수 있다. 그러므로 호스트는 후기를 작성하는 고객들이 좋은 후기를 남길 수 있도록 가장 중요시 여기는 청결을 관리해 고객들을 만족시킬 필요가 있다.

### 안 좋은 후기 분석 결과
- 워드클라우드 분석 결과 '너무, 그리고, 하지만, 굉장히' 등의 접속사나 부사를 제외하고 가장 많이 언급된 단어는 '체크인, 부엌, 화장실이, 냄새가, 물이' 이다. 
- 체크인의 경우 체크인에 대한 불만을 말한 것이 아닌 단순히 체크인 시간대에 대한 언급이었으므로 중요하다고 볼 수 없다. '부엌'에 대해서는 식기나 도구가 더럽다는 불만이 많았다.  '화장실이'에 대해서는 좁다, 춥다, 더럽다, 냄새 난다 등의 언급이 많았다. '냄새가'의 경우 욕실이나 수건 등에서 냄새가 난다는 것이었으므로 화장실과 연관되었다고 볼 수 있다. '물이'의 경우도 '화장실'의 물이 잘 안 내려간다거나 차갑다는 등의 내용이었으므로 마찬가지로 화장실에 대한 불만으로 볼 수 있다. 

### 안 좋은 후기 시사점
- 이 분석 결과를 청결도에 대한 만족도와 연관 지어 설명할 수 있다. 특히 화장실 청결상태나 냄새, 시설에 대한 불만이 많은 것을 볼 수 있는데, 이는 청결도에 대한 만족도에 화장실의 청결상태가 많은 영향을 미친다고 해석할 수 있다. 
- 따라서 호스트들은 화장실을 항상 깨끗한 상태로 유지시키기 위해 더 많은 노력을 기울여야 한다. 더 나아가 부엌의 청결도에도 신경을 써서 게스트가 머물기 전 주방 식기와 도구를 항상 깨끗이 씻고 관리하는 것도 중요하다. 
- 전반적으로 이 분석을 통해서 에어비앤비 숙소 호스트는 안 좋은 후기를 피하기 위해 청결도를 많이 신경 써야 함을 알 수 있다.

## 6. 시사점

 위의 회귀 분석에 따르면, 에어비앤비 사용 고객들의 만족도에 가장 영향을 미치는 변수들로 청결도, 후기, 호스트 응답률이 나왔다. 에어비앤비가 현재 고객 수의 유지와 보다 더 많은 고객들을 모으기 위해서는, 모든 변수들에 노력을 기하기 보다는 가장 영향력 있던 세가지 변수에 선택과 집중을 하는 것이 더 효율적 일 것이다. 각각의 변수마다 조언을 하자면:
1. 청결도 – 청결도는 높으면 높을수록 고객의 만족도가 올라가는 데 큰 영향력이 있었다. 따라서 에어비앤비와 호스트는 자신의 집이나 방을 항상 청결하게 유지하여 고객에게 신뢰를 주어야 한다. 이것은 청소를 외주업체에 의뢰하여, 장기적 체결로 청소비용을 할인 받으며, 대신 항상 청결을 유지하는 것을 보장받아야 한다. 간혹 호스트가 직접 방이나 집을 청소 하던 중, 깜빡 하고 청소를 하지 못하거나 미처 다 치우지 못할 때가 있던 상황을 사전에 방지할 수 있게 될 것이다. 비록 당장에는 청소비용이 부가로 나가는 것 같아도, 장기적으로 봤을 때 고객을 유지하고 지속적 고객 방문으로 인해 일정한 수입으로도 이어질 수 있으니, 크게 보면 더욱 유리할 것이라 생각된다. 

2. 후기 – 에어비앤비 사용 후 고객들의 긍정적인 리뷰들을 보면, 대게 호스트가 친절했다거나 그가 제공했던 좋은 서비스들의 예가 있다. 호스트의 특별한 서비스가 그 집만의 특이점으로, 한 번 이용해봤던 고객들은 좋은 추억을 만들 수 있고, 또 그들이 그들 지인들에게 기꺼이 추천을 하여 더 많은 고객 유치로 이어질 수 있게 된다. 우리나라 호스트들이 했던 서비스의 예로는 한복 대여 서비스나 친절하게 직접 만든 맛집/쇼핑/관광 지도 (현지인들만 아는 곳들로)가 있었다. 이런 식으로 에어비앤비에서 주도하여 호스트마다 특별한 서비스를 제공하도록 하거나, 호스트 본인들이 고객을 더 모으기 위해 참고하여 제공하면 좋을 것 같다. 

3. 호스트 응답률 – 설문조사 후, 가장 높은 점수로 고객의 만족도에 큰 영향력을 끼친 변수로 호스트 응답률이 거의 100%로 이어졌다. 실제로 주변 지인들이 외국에 나가서 에어비앤비를 사용했을 때, 호스트와의 의사소통이 이루어지지 않아서 여행에 차질이 생기는 사례가 많았다. 이 경우, 보통은 호스트가 한 명이고 별다른 연락수단 방법이 없어서 생긴 것이다. 따라서, 불의의 사건, 사고의 발생으로 호스트 본인이 연락이 안될 시, 본인을 제외한 한 명을 더 비상연락망으로 제공하면 좋을 것 같다. 그로써, 고객이 좀 더 호스트에게 빠른 응답을 받을 수 있을 것으로 기대된다.
